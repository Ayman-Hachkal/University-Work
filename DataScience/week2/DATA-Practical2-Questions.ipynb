### **What do the reviewers think about the quality of wines?**


#### **T7) Calculate the metrics of central tendency for wine ratings (points)**


```python
#Write your answer here
points_median = df['points'].median()
points_mean = df['points'].mean()
points_mode = df['points'].mode()

print("Mean:    ", points_mean)
print("Median:  ", points_median)
print("Mode:    ", points_mode)
```

#### **T8) Calculate the metrics of dispersion for wine ratings (points)**


```python
#Write your answer here
points_min = df['points'].min()
points_max = df['points'].max()
points_range = points_max - points_min
points_std = df['points'].std()

print("Points Min:      ", points_min)
print("Points Max:      ", points_max)
print("Points Range:    ", points_range)
print("Points STD:      ", points_std)
```

#### **T9) Calculate the interquartile range for the ratings of all reviewed wines**


```python
#Write your answer here
points_q1 = np.percentile(df['points'], 25) 
points_q3 = np.percentile(df['points'], 75)
points_iqr = points_q3 - points_q1
print("Q1:  ", points_q1)
print("Q3:  ", points_q3)
print("IQR: ", points_iqr)
```

#### **T12) Which is the most reviewed wine variety and what is its average rating?**

* Hint: Check the section on array masking from the Pandas/NumPy chapters in the Python Data Science Handbook


```python
#Write your answer here
variety_mode = df.variety.mode()
print(variety_mode.to_string())
variety_mode_mean = df.mask(df['variety'] == variety_mode.to_string()).points.mean()
print(variety_mode_mean)
```

#### **T13) Which are the most widely reviewed wineries? How many reviews did each receive?**

* Hint: Another option is to use the function argwhere from NumPy
