```python
import sqlite3
import pandas as pd


db_file = "example.db"

# Name of table to create
passenger_table_name = "passenger"
cabin_inventory_table = "cabin_inventory"


def create_database(db_name):
    conn = sqlite3.connect(db_name)
    conn.execute("PRAGMA foreign_keys = ON")
    return conn


def upload_datasets(conn, passenger_file, cabin_file):

    passenger_df = pd.read_csv(passenger_file)
    cabin_df = pd.read_csv(cabin_file)

    # Upload cabin table first (parent table)
    cabin_df.to_sql(cabin_inventory_table, conn, if_exists="replace", index=False)

    # Upload passenger table
    passenger_df.to_sql(passenger_table_name, conn, if_exists="replace", index=False)

    print("Datasets successfully uploaded.")


def show_tables(conn):
    print("\nPassengers Table")
    passengers = pd.read_sql("SELECT * FROM passenger", conn)
    print(passengers.head())

    print("\nCabin Inventory Table")
    cabins = pd.read_sql("SELECT * FROM cabin_inventory", conn)
    print(cabins.head())

def join_passenger_cabin(conn):
    query = """
            SELECT
                p.personal_name,
                p.family_name,
                c.class, 
                c.ship_level, 
                c.price
            FROM passenger p
            JOIN cabin_inventory c
            ON p.cabin_id = c.cabin_id"""
        
    df = pd.read_sql(query, conn)
    print("\nPassenger Cabin Details")
    print(df)

def passenger_by_cabin(conn):
    query = """
        SELECT 
            c.cabin_id,
            c.class,
            p.personal_name,
            p.family_name
        FROM cabin_inventory c
        JOIN passenger p
        ON c.cabin_id = p.cabin_id
        ORDER BY c.cabin_id"""

    df = pd.read_sql(query, conn)
    print("\nPassengers by Cabin")
    print(df)


def main():

    passenger_file = "reduced_lusitania_simple_join_dataset.csv"  #<< names of the .csv files
    cabin_file = "reduced_cabin_inventory.csv"

    conn = create_database(db_file)     #connects to the database

    upload_datasets(conn, passenger_file, cabin_file)  #<< loads the .csv files

    #show_tables(conn)  #<< shows the tables of the .csv files

    join_passenger_cabin(conn)
    passenger_by_cabin(conn)
   
    #Add other function calls below

    conn.close()

if __name__ == "__main__":
    main()
```
