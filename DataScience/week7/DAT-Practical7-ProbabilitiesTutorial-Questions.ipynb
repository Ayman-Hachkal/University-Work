The main methods for continuous variables (like normal and exponential) are:
The main public methods for continuous RVs are:

* **scipy.stats.norm.rvs(...)**:  generate the random variates for the specified distribution
* **scipy.stats.norm.pdf(...)**:  generate the Probability Distribution Function (PDF) values for the given observations
* **scipy.stats.norm.cdf(...)**:  generate the Cumulative Distribution Function (PDF) values for the given observations, e.g., what is the probability of finding values less than or equal to X?
* **scipy.stats.norm.ppf(...)**:  given a percent point value Y%, what is the value X that would be larger than Y% of the data
