```python
# Column vectors
x = np.array([[2],
              [3],
              [5]])

v = np.array([[3],
              [0],
              [4]])

# x^T v  (1×3 @ 3×1 → scalar 1×1)
xtv = x.T @ v         
print(xtv)              

# x v^T  (3×1 @ 1×3 → 3×3 matrix)
xvt = x @ v.T
print(xvt)
```

```python
A = np.array([[4, 1, 0, 2],
              [2, 0, 1, 3]])

b = np.array([[3],
              [0],
              [5],
              [1]])

print(A.shape) 
print(b.shape) 

c = A @ b    
print(c)

```

```python
# 2c

def dot(v1, v2):
    """
    Compute the dot product of two NumPy arrays v1 and v2
    without using np.dot or np.matmul.
    """

    total = 0.0
    for a, b in zip(v1, v2):
        total += a * b
    return total
```

```python
import numpy as np

x = np.array([2, 3, 5])
v = np.array([3, 0, 4])

print(dot(x, v))
```

```python
#2e correct
def transform(data):
    return np.dot(data[:, 1:3], np.array([[1], [1]]))

data = np.array([
    [1, 17,  9],
    [2, 21, 10],
    [3, 12, 17]
])

print(transform(data))
```
